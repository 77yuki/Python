```python
import numpy as np
import matplotlib.pyplot as plt

T = 2
paths = 70000
N = [100, 200, 400, 800, 1600, 3200, 6400]


mean = np.zeros(len(N))
variance = np.zeros(len(N))
deltat = [T/100, T/200, T/400, T/800, T/1600, T/3200, T/6400]

for j in range(len(N)):
    dt = deltat[j]
    I = np.zeros(paths)
    for i in range(N[j]):
        phi_t_i = np.random.normal(0, 1, paths)
        delta_z_i = phi_t_i * np.sqrt(dt)
        I += delta_z_i **2
    mean[j] = np.mean(I)
    variance[j] = np.var(I)
    
print('dt\t\tmean\t\t\tvariance') #This is the heading of the table

for i in np.arange(7):
    if len(str(deltat[i])) < 6:
        print (str(deltat[i]) + ": \t\t" + str(mean[i]) + "\t" + str(variance[i])) 
    else:
        print (str(deltat[i]) + ": \t" + str(mean[i]) + "\t" + str(variance[i]))   

           #plot a graph
plt.plot(deltat, variance)
plt.xlabel('deltat')
plt.ylabel('var(I(T, deltat))')
```

By the text-based table of mean and variance of I(T, $\Delta t$) for the different values of $\Delta t$, we find that mean is around 2, which is the same as the value of T, and variance is decreasing and goes to 0 as $\Delta t$ decreases. By observing the plot, we find that as $\Delta t$ increases, the variance of I(T, $\Delta t$) increases. Also, we observe that the relationship between $\Delta t$ and I(T, $\Delta t$) is positive linear.
